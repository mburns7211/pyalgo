````python
import os
import random
import numpy as np
from pylab import mpl, plt
plt.style.use('seaborn-v0_8')
mpl.rcParams['savefig.dpi'] = 300
mpl.rcParams['font.family'] = 'serif'
os.environ['PYTHONHASHSEED'] = '0'
````

````python
# evenly spaced grid of floats from 0 to 10
x = np.linspace(0,10)
````

````python
def set_seeds(seed=100):
    random.seed(seed)
    np.random.seed(seed)
````

````python
set_seeds()
````

````python
# generates random data for y
y = x + np.random.standard_normal(len(x))
````

````python
# OLS regression of degree 1 (linear regression)
reg = np.polyfit(x, y, deg=1)
````

````python
# results in an array of tuples [(val1, val2)]
# for 1st degree function this means y = val1 * x + val2 (slope, intercept)
reg
````

````python
plt.figure(figsize=(10,6))
plt.plot(x, y, 'bo', label='data')
plt.plot(x, np.polyval(reg,x), 'r', lw=2.5, label='linear regression')
plt.legend(loc=0)
````

````python
# expanding the interval allows us to predict values for y beyond the domain of the original data set through extrapolation
plt.figure(figsize=(10,6))
plt.plot(x, y, 'bo', label='data')
xn = np.linspace(0,20)
plt.plot(xn, np.polyval(reg,xn), 'r', lw=2.5, label='linear regression')
plt.legend(loc=0)
````

## Basic Idea for price prediction

```
The above example uses linear regression on a random dataset of values.  
The order of the values does not matter (as long as x and y pairs are constent)
For price predicting we predict tomorrows price based on today, or with the day before etc.  
Each day before used is known as lags
```

````python
x = np.arange(12)
````

````python
x
````

```
3 lags - 3 independent variables and 1 dependent
0, 1, 2 are independent the 3 is corresponding value for dependent variable
move forward 1,2,3 independent 4 dependent etc
Linear equation : A (DOT) x = b
A: matrix
x and b are vectors
```

````python
lags = 3
m = np.zeros((lags+1, len(x) - lags))

m[lags] = x[lags:]
for i in range(lags):
    m[i] = x[i:i - lags]
````

````python
m.T
````

````python
# implement linear ols regression
reg = np.linalg.lstsq(m[:lags].T, m[lags], rcond=None)[0] 
````

````python
reg
````

````python
# applies regression results over rest of matrix using dot product, resulting in correct values
np.dot(m[:lags].T, reg)
````
