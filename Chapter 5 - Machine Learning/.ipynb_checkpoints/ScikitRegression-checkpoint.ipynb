```python
import numpy as np
```

```python
# We will revist the simple prediction of next number now using sklearn
```

```python
x = np.arange(12)
```

```python
x
```

```python
lags = 3
```

```python
m = np.zeros((lags + 1, len(x) - lags))
```

```python
m[lags] = x[lags:]
```

```python
for i in range(lags):
    m[i] = x[i:i - lags]
```

```python
m
```

```python
# Use Scikit learn for this purpose requires 3 steps:
# 1. Model Selection - a model is to be picked and instantiated
# 2. Model Fitting - the models is to be fitted to the data at hand
# 3. Prediction - given the fitted model, the prediction is conducted
```

```python
from sklearn import linear_model
```

```python
lm = linear_model.LinearRegression()
```

```python
lm.fit(m[:lags].T, m[lags])
```

```python
lm.coef_
```

```python
lm.intercept_
```

```python
lm.predict(m[:lags].T)
```

```python
# force a fit without an intercept value
lm = linear_model.LinearRegression(fit_intercept=False)
```

```python
lm.fit(m[:lags].T, m[lags])
```

```python
lm.coef_
```

```python
lm.intercept_
```

```python
lm.predict(m[:lags].T)
```
