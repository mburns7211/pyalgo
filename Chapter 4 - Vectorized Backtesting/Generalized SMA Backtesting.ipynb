```python
import SMAVectorBacktester as SMA
```

```python
smabt = SMA.SMAVectorBacktester(
    symbol='EUR=',
    SMA1=42,
    SMA2=252,
    start='2010-1-1',
    end='2019-12-31'
)
```

```python
# returns gross performance, under performance of the strategy
smabt.run_strategy()
```

```python
%%time
# SMA1 and SMA2 tuples in form of
# (start, end, step size)
x = smabt.optimize_parameters(
    (30, 50, 2),
    (200, 300, 2)
)
print(f"SMA1 best val: {x[0][0]}, SMA2 best val:{x[0][1]}, strategy return {x[1] * 100}%")
```

```python
smabt.plot_results()
```
